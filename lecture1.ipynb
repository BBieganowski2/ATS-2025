```python
import numpy as np
```

```python
class YourTeamName:

    def get_move(self, my_moves, opp_moves):
        return np.random.choice(['rock', 'paper', 'scissors'])

class OpponentTeamName:

    def get_move(self, my_moves, opp_moves):
        return np.random.choice(['rock', 'paper', 'scissors'])
```

```python
your_score = 0
opponent_score = 0

yourmoves = []
opponent_moves = []

yourteam = YourTeamName()
opponent = OpponentTeamName()

for i in range(10_000):
    your_move = yourteam.get_move(yourmoves, opponent_moves)
    opponent_move = opponent.get_move(opponent_moves, yourmoves)

    yourmoves.append(your_move)
    opponent_moves.append(opponent_move) 

    if your_move == opponent_move:
        winner = np.random.choice(['you', 'opponent'])
        if winner == 'you':
            your_score += 1
        else:
            opponent_score += 1
    elif (your_move == 'rock' and opponent_move == 'scissors') or \
         (your_move == 'paper' and opponent_move == 'rock') or \
         (your_move == 'scissors' and opponent_move == 'paper'):
        your_score += 1
    else:
        opponent_score += 1
```

```python
import numpy as np
import pandas as pd

def market_metrics(bid, bidv, ask, askv, side, volume):
    
    #  mid-price
    mid_price = (bid[0] + ask[0]) / 2
    
    # normal spread
    normal_spread = ask[0] - bid[0]
    
    # Spread as percentage of mid-price
    spread_percentage = (normal_spread / mid_price) * 100
    
    # effective spread based on market order side
    def get_effective_spread(order_side, order_volume):
        remaining_volume = order_volume
        cost = 0
        
        if order_side == 'bid':  # Market buy order, takes liquidity from ask side
            for price, vol in zip(ask, askv):
                if remaining_volume > vol:
                    cost += price * vol
                    remaining_volume -= vol
                else:
                    cost += price * remaining_volume
                    break
        else:  # Market sell order, takes liquidity from bid side
            for price, vol in zip(bid, bidv):
                if remaining_volume > vol:
                    cost += price * vol
                    remaining_volume -= vol
                else:
                    cost += price * remaining_volume
                    break
        
        avg_execution_price = cost / order_volume
        effective_spread = 2 * abs(avg_execution_price - mid_price)
        effective_spread_percentage = (effective_spread / mid_price) * 100
        
        return effective_spread_percentage
    
    effective_spread_percentage = get_effective_spread(side, volume)
    
    return {
        "Mid Price": mid_price,
        "Normal Spread": normal_spread,
        "Spread % of Mid Price": spread_percentage,
        "Effective Spread %": effective_spread_percentage
    }

# Example:
bid_prices = [99, 98.5, 98]
bid_volumes = [50, 100, 150]
ask_prices = [100, 100.5, 101]
ask_volumes = [60, 110, 140]
side = 'bid'  # Market buy order
market_order_size = 80

metrics = market_metrics(bid_prices, bid_volumes, ask_prices, ask_volumes, side, market_order_size)
print(metrics)
```
